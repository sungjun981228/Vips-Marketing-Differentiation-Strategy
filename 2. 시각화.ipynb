```python
import matplotlib.pyplot as plt

#  
brands = ['', '', '', '', '', '', '', '', '']

#  
x_raw = [49700, 27900, 33900, 34900, 29900, 21900, 11900, 17900, 46900]  # 
y_raw = [61155, 156652, 139818, 15162, 2802, 118759, 212578, 7970, 1973]  # 

#   (1~10 )
def normalize_to_1_10(values):
    min_val = min(values)
    max_val = max(values)
    return [round(1 + (v - min_val) / (max_val - min_val) * 9, 2) for v in values]

#  
x = normalize_to_1_10(x_raw)
y = normalize_to_1_10(y_raw)

# 
plt.figure(figsize=(10, 7))
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'teal', 'brown']

for i in range(len(brands)):
    plt.scatter(x[i], y[i], color=colors[i], s=200)
    plt.text(x[i] + 0.2, y[i], brands[i], fontsize=11)

#  
plt.xlim(0, 11)
plt.ylim(0, 11)
plt.xlabel('  (  )', fontsize=12)
plt.ylabel(' (  )', fontsize=12)
plt.title('   ( vs )', fontsize=14)
plt.grid(True)
plt.axhline(y=5, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(x=5, color='gray', linestyle='--', linewidth=0.8)
plt.tight_layout()
plt.show()
```
