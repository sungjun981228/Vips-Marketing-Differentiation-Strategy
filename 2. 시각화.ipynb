```python
import matplotlib.pyplot as plt

# 브랜드 이름
brands = ['빕스', '애슐리', '쿠우쿠우', '고메스퀘어', '자연별곡', '명륜진사갈비', '두끼', '피자몰', '드마리스']

# 원본 데이터
x_raw = [49700, 27900, 33900, 34900, 29900, 21900, 11900, 17900, 46900]  # 가격
y_raw = [61155, 156652, 139818, 15162, 2802, 118759, 212578, 7970, 1973]  # 언급량

# 정규화 함수 (1~10 범위)
def normalize_to_1_10(values):
    min_val = min(values)
    max_val = max(values)
    return [round(1 + (v - min_val) / (max_val - min_val) * 9, 2) for v in values]

# 정규화된 값
x = normalize_to_1_10(x_raw)
y = normalize_to_1_10(y_raw)

# 시각화
plt.figure(figsize=(10, 7))
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'teal', 'brown']

for i in range(len(brands)):
    plt.scatter(x[i], y[i], color=colors[i], s=200)
    plt.text(x[i] + 0.2, y[i], brands[i], fontsize=11)

# 축 설정
plt.xlim(0, 11)
plt.ylim(0, 11)
plt.xlabel('가격 포지션 (가성비 → 프리미엄)', fontsize=12)
plt.ylabel('언급량 (낮음 → 높음)', fontsize=12)
plt.title('브랜드 포지셔닝 맵 (가격 vs 언급량)', fontsize=14)
plt.grid(True)
plt.axhline(y=5, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(x=5, color='gray', linestyle='--', linewidth=0.8)
plt.tight_layout()
plt.show()
```
